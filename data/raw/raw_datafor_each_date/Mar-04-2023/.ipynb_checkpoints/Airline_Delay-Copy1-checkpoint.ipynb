```python
time_str = "2h 52m"

# Split the string into hours and minutes
split_time = time_str.split()
hours = int(split_time[0][0]) if 'h' in split_time[0] else 0
minutes = int(split_time[1][0]) if 'm' in split_time[1] else 0

# Convert hours to minutes and add to the total
total_minutes = hours * 60 + minutes

print(f"The total minutes for {time_str} is: {total_minutes} minutes")
```

```python
import pandas as pd

# Create a DataFrame
df = pd.read_csv('Airline-France-320.csv')

# Function to convert time strings to minutes
def time_to_minutes(time_str):
    parts = time_str.split()
    total_minutes = 0
    for part in parts:
        if 'h' in part:
            total_minutes += int(part[:-1]) * 60
        elif 'm' in part:
            total_minutes += int(part[:-1])
    return total_minutes

# Apply the function to the "RealArrivalTime" column
df['RealArrivalTime'] = df['RealArrivalTime'].apply(time_to_minutes)

# Print the updated DataFrame
print(df)
```
