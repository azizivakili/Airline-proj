```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Sample data
data = pd.read_csv("Mar-31-today.csv")

# Create a DataFrame
df = pd.DataFrame(data)

```

```python
# Create a DataFrame
df = pd.DataFrame(data)

# Convert string times to datetime objects
df['RealArrivalTime'] = pd.to_datetime(df['Estimated Arrival Time'], format='%H:%M:%S')

# Function to add random minutes between -20 to +50 minutes
def add_random_minutes(time):
    random_minutes = np.random.randint(-20, 51)  # Random integer between -20 to +50
    new_time = time + timedelta(minutes=random_minutes)
    return new_time

# Apply the function to the 'RealArrivalTime' column
df['RealArrivalTime'] = df['RealArrivalTime'].apply(add_random_minutes)

# Function to convert time to "Xh Ymin" format
def convert_to_hours_minutes(time):
    hours = time.hour
    minutes = time.minute
    return f"{hours}h {minutes}m"

# Apply the function to the 'RealArrivalTime' column
df['RealArrivalTime'] = df['RealArrivalTime'].apply(convert_to_hours_minutes)

# Print the DataFrame
print(df)
```

```python
df['RealArrivalTime'].replace('nanh nanm', '', inplace=True)
```
