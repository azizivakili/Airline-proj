```python
import pandas as pd

# Your provided data
data = pd.read_csv("Mar-31-data.csv")
print(data)
# Create a DataFrame
df = pd.DataFrame(data)

# Convert "Estimated Arrival Time" to minutes
df['Estimated Arrival Time'] = pd.to_timedelta(df['Estimated Arrival Time'])
df['EA_Time'] = df['Estimated Arrival Time'].dt.total_seconds() / 60
# Print the result DataFrame
#print(df)

df.to_csv("EA_Time.csv")
```

```python
# Convert RealArrivalTime to Minutes. Eg: 3h 01m ->  181 min

def time_to_minutes(time_str):
    if pd.isna(time_str):
        return None
    else:
        hours, minutes = time_str.split('h ')
        hours = int(hours)
        minutes = int(minutes[:-1])  # Removing the 'm' from minutes
        return hours * 60 + minutes

# Apply the function to each row in the DataFrame
df['RA_Time'] = df['RealArrivalTime'].apply(time_to_minutes)
#df
```

```python
df.to_csv("RA_Time.csv")
```

```python
# Find Dealy_Time in minutes

df['Delay_Time'] = df['RA_Time'] - df['EA_Time']
```

```python
# Extract Departure_Date form Departure Time

def convert_day_to_date(day):
    if day == 'Mon':
        return '04.03.2024'
    elif day == 'Tue':
        return '05.03.2024'
    elif day == 'Wed':
        return '06.03.2024'
    elif day == 'Thu':
        return '07.03.2024'
    elif day == 'Fri':
        return '08.03.2024'
    elif day == 'Sat':
        return '09.03.2024'
    elif day == 'Sun':
        return '10.03.2024'
    
df['Departure_Date'] = df['Departure Time'].str.split().str[0].apply(convert_day_to_date)
#df.to_csv("extract_dates.csv")
#df
```

```python
# Extract Arrival_Date form Arrival Time

def convert_day_to_date(day):
    if day == 'Mon':
        return '04.03.2024'
    elif day == 'Tue':
        return '05.03.2024'
    elif day == 'Wed':
        return '06.03.2024'
    elif day == 'Thu':
        return '07.03.2024'
    elif day == 'Fri':
        return '08.03.2024'
    elif day == 'Sat':
        return '09.03.2024'
    elif day == 'Sun':
        return '10.03.2024'
    
df['Arrival_Date'] = df['Arrival Time'].str.split().str[0].apply(convert_day_to_date)
#df.to_csv("extract_dates.csv")
df.head()
```

```python
# Clean time in format of HH:MM AM/PM

df['Arrival_Time'] = df['Arrival Time'].str.extract(r'(\d{2}:\d{2}[AP]M)')
df['Departure_Time'] = df['Departure Time'].str.extract(r'(\d{2}:\d{2}[AP]M)')
```

```python
# Clean time in format of 24 hours


df['Arrival_Time'] = pd.to_datetime(df['Arrival_Time']).dt.strftime('%H:%M')
df['Departure_Time'] = pd.to_datetime(df['Departure_Time']).dt.strftime('%H:%M')
df.head(3)
```

```python
df = df.drop(columns=['Departure Time', 'Arrival Time'])
df.head()
```

```python
# Re-structure the columns 

df = df[['Aircraft','Flight Type','Origin','Destination', 'Departure_Date', 'Departure_Time', 'Arrival_Date', 'Arrival_Time', 'EA_Time', 'RA_Time', 'Delay_Time']]
df
```

```python
df.to_csv("Mar-31-Final_DataSet.csv")
```
