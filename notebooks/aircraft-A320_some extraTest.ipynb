```python
headers = [header.text.strip() for header in rows[1].find_all('th')]
headers = headers[1:]
headers
```

```python
data = []
for row in rows[4:]:
    data.append([td.text.strip() for td in row.find_all('td')])
    
#columns = ['Flight', 'Aircraft', 'Origin', 'Destination', 'Departure', 'Estimated Arrival', 'Duration']
```

```python
df = pd.DataFrame(data, columns=headers)
```

```python
### Data Cleaning 
```

```python
df.rename(columns={'Ident': 'Aircraft_name'}, inplace=True)
```

```python
# remove the first index column
#df = df.drop(columns = 'index')
df
```

```python
#df = df.drop(columns = ['A_day' , 'A_hour'])
df['Departure'].str.split(',' , 1 , expand = True )
```

```python
df[['Arr_DayOfWeek', 'Arr_Time']] = df['EstimatedArrival Time'].str.extract(r'(\w{3}) (\d{2}:\d{2}[APMapm]{2})')
```

```python
df
```

```python
df[['Dep_DayOfWeek', 'Dep_Time']] = df['EstimatedArrival Time'].str.extract(r'(\w{3}) (\d{2}:\d{2}[APMapm]{2})')
```

```python
df[['Arr_DayOfWeek', 'Arr_Time']] = df['EstimatedArrival Time'].str.extract(r'(\w{3}) (\d{2}:\d{2}[APMapm]{2})')
```

```python
df
```

```python

df = df.drop(columns = 'Arr_Time' )
df
```

```python
#all the cell when there is not value return to blank

#df = df.fillna("")
#df1['Paying Customer'] = df1['Paying Customer'].str.replace('No' , 'N')
#df['Dep_DayOfWeek'] = df['Dep_DayOfWeek'].str.replace('NaN' , 'Wed')

#-----------------------------------------------------------------
df['Dep_DayOfWeek'] = df['Dep_DayOfWeek'].str.replace('NaN', 'Wed')

# Using fillna to replace actual NaN values with 'Wed'
df['Dep_DayOfWeek'] = df['Dep_DayOfWeek'].fillna('Wed')
```

```python
df['Dep_Time'] = df['Dep_Time'].str.replace('NaN', '11:29PM ')

# Using fillna to replace actual NaN values with 'Wed'
df['Dep_Time'] = df['Dep_Time'].fillna('11:29PM ')

#----------------------------------------------------------
df['Arr_DayOfWeek'] = df['Arr_DayOfWeek'].str.replace('NaN', '11:29PM ')

# Using fillna to replace actual NaN values with 'Wed'
df['Arr_DayOfWeek'] = df['Arr_DayOfWeek'].fillna('11:29PM ')
```
